# **Cluster Analysis and Interpretation**

###  **K-Means Clustering (k=4)**

* **Nature**: Partition-based, centroid-driven.
* **Observation**: Created compact, spherical clusters based on Euclidean distance.
* **Insights**:

  * Cluster 0: High `Balance` and `Bonus_miles`  loyal frequent flyers.
  * Cluster 1: Low `Flight_miles_12mo` and recent `Days_since_enroll`  new/less active users.
  * Cluster 2: Moderate values across metrics  balanced user behavior.
  * Cluster 3: Low `Balance`, high `Qual_miles`  possibly business travelers.



### **Hierarchical Clustering (Agglomerative  Ward Linkage)**

* **Nature**: Tree-based, builds nested clusters.
* **Observation**: Dendrogram showed meaningful splits at 35 clusters.
* **Insights**:

  * Clear hierarchy between high and low mileage users.
  * Captured subtle variations, e.g., users with similar `Flight_trans_12` but different `cc_miles`.
  * Suitable for discovering structured subgroups.


###  **DBSCAN (Density-Based)**

* **Nature**: Density-driven, handles noise/outliers.
* **Observation**: Formed arbitrary-shaped clusters and isolated sparse users as noise.
* **Insights**:

  * Identified dense regions of frequent flyers.
  * Labeled outliers with low `Flight_miles_12mo` and `cc_miles` as noise.
  * Best for detecting non-spherical clusters and filtering inactive or anomalous profiles.

```python
# Assuming you already have dbscan_labels from DBSCAN
# Filter out noise points (-1) because silhouette score doesn't support them
mask = dbscan_labels != -1
if len(set(dbscan_labels)) > 1 and sum(mask) > 1:
    dbscan_score = silhouette_score(numeric_df[mask], dbscan_labels[mask])
    print(f"Silhouette Score for DBSCAN (excluding noise): {dbscan_score:.3f}")
else:
    print("Silhouette Score for DBSCAN cannot be calculated (too few clusters or all noise).")
```

Silhouette Score Range: [-1, 1]

+1  Perfect clustering

0  Overlapping clusters

< 0  Wrong clustering (samples assigned to wrong clusters)

```python
import matplotlib.pyplot as plt

# Assuming these scores are already calculated:
# kmeans_score, dbscan_score (with mask), hier_score

# Handle DBSCAN noise
dbscan_valid = dbscan_labels != -1
if len(set(dbscan_labels)) > 1 and sum(dbscan_valid) > 1:
    dbscan_score = silhouette_score(numeric_df[dbscan_valid], dbscan_labels[dbscan_valid])
else:
    dbscan_score = None

# Prepare data for plot
scores = {
    'K-Means': kmeans_score,
    'DBSCAN': dbscan_score if dbscan_score is not None else 0,
    'Hierarchical': hier_score
}

# Bar Plot
plt.figure(figsize=(8, 5))
bars = plt.bar(scores.keys(), scores.values(), color=['skyblue', 'salmon', 'lightgreen'])
plt.ylim(0, 1)
plt.ylabel('Silhouette Score')
plt.title('Clustering Quality Comparison')

# Annotate scores
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height + 0.02, f'{height:.2f}', ha='center')

plt.show()
```

Based on the internal evaluation using **Silhouette Score, K-Means** and **Hierarchical Clustering** demonstrate better intra-cluster cohesion and inter-cluster separation compared to **DBSCAN**. While K-Means achieves a higher score, indicating compact and well-separated clusters in the feature space, DBSCAN may yield lower scores due to the presence of noise points (label = -1) and sensitivity to parameter tuning (eps, min_samples). Overall,** K-Means** is more effective for this dataset assuming spherical cluster structure, while **DBSCAN** is better suited for identifying arbitrary-shaped clusters and outliers.

K-Means, DBSCAN, and Hierarchical clustering were applied to a scaled dataset and visualized using PCA. Based on Silhouette Scores, K-Means achieved the best clustering quality (score = 0.71), forming compact and well-separated clusters. DBSCAN showed lower performance due to noise sensitivity (score = 0.41), while Hierarchical Clustering performed reasonably well (score = 0.66).
