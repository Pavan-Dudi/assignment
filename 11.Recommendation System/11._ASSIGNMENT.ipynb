```python
# prompt: Compute similarity

from sklearn.metrics.pairwise import cosine_similarity


# Get all one-hot encoded columns and scaled numerical columns
features_for_similarity = [col for col in df.columns if col.startswith(('genre_', 'type_')) or col in ['rating', 'episodes']]

similarity_matrix = cosine_similarity(df[features_for_similarity])
```

```python
# prompt: Given a target anime, recommend a list of similar anime based on cosine similarity scores.

import pandas as pd
import numpy as np
def recommend_similar_anime(target_anime_name, df, similarity_matrix, n=10):
  """
    Given a target anime, recommends a list of similar anime based on cosine similarity scores.

    Args:
      target_anime_name (str): The name of the target anime.
      df (pd.DataFrame): The DataFrame containing anime data.
      similarity_matrix (np.ndarray): The precomputed cosine similarity matrix.
      n (int): The number of recommendations to return.

    Returns:
      pd.DataFrame: A DataFrame containing the recommended anime names and their similarity scores.
  """
  # Get the index of the target anime
  target_anime_index = df[df['name'] == target_anime_name].index[0]

  # Get the similarity scores for the target anime
  similarity_scores = list(enumerate(similarity_matrix[target_anime_index]))

  # Sort the anime based on similarity scores in descending order
  sorted_similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

  # Get the indices of the top n similar anime (excluding the target anime itself)
  top_n_anime_indices = [i[0] for i in sorted_similarity_scores[1:n+1]]

  # Get the names and similarity scores of the top n similar anime
  recommended_anime = df.iloc[top_n_anime_indices][['name']]
  recommended_anime['similarity_score'] = [score for index, score in sorted_similarity_scores[1:n+1]]

  return recommended_anime

# Example usage:
# Let's recommend anime similar to 'Naruto'
target_anime_name = 'Naruto'
recommended_anime = recommend_similar_anime(target_anime_name, df, similarity_matrix)
print(f"Recommendations for '{target_anime_name}':")
recommended_anime
```

```python
# prompt: Experiment with different threshold values for similarity scores to adjust the recommendation list size

def recommend_similar_anime_with_threshold(target_anime_name, df, similarity_matrix, similarity_threshold=0.8):
  """
    Given a target anime, recommends a list of similar anime based on cosine similarity scores,
    filtering by a similarity threshold.

    Args:
      target_anime_name (str): The name of the target anime.
      df (pd.DataFrame): The DataFrame containing anime data.
      similarity_matrix (np.ndarray): The precomputed cosine similarity matrix.
      similarity_threshold (float): The minimum similarity score for an anime to be included in recommendations.

    Returns:
      pd.DataFrame: A DataFrame containing the recommended anime names and their similarity scores,
                    filtered by the threshold.
  """
  # Get the index of the target anime
  try:
    target_anime_index = df[df['name'] == target_anime_name].index[0]
  except IndexError:
    print(f"Anime '{target_anime_name}' not found in the dataset.")
    return pd.DataFrame(columns=['name', 'similarity_score'])

  # Get the similarity scores for the target anime
  similarity_scores = list(enumerate(similarity_matrix[target_anime_index]))

  # Filter anime based on the similarity threshold and exclude the target anime itself
  filtered_recommendations = [(i, score) for i, score in similarity_scores if i != target_anime_index and score >= similarity_threshold]

  # Sort the filtered anime based on similarity scores in descending order
  sorted_filtered_recommendations = sorted(filtered_recommendations, key=lambda x: x[1], reverse=True)

  # Get the names and similarity scores of the recommended anime
  recommended_anime_indices = [i[0] for i in sorted_filtered_recommendations]
  recommended_anime_scores = [score for index, score in sorted_filtered_recommendations]

  if not recommended_anime_indices:
      print(f"No recommendations found for '{target_anime_name}' with a similarity threshold of {similarity_threshold}.")
      return pd.DataFrame(columns=['name', 'similarity_score'])

  recommended_anime_df = df.iloc[recommended_anime_indices][['name']].copy()
  recommended_anime_df['similarity_score'] = recommended_anime_scores

  return recommended_anime_df

# Experiment with different threshold values
target_anime_name = 'Naruto'
threshold_values = [0.5, 0.6, 0.7, 0.8, 0.9]

for threshold in threshold_values:
  print(f"\nRecommendations for '{target_anime_name}' with similarity threshold = {threshold}:")
  recommended_anime = recommend_similar_anime_with_threshold(target_anime_name, df, similarity_matrix, similarity_threshold=threshold)
  print(recommended_anime)
  print(f"Number of recommendations: {len(recommended_anime)}")
```

```python
# prompt: Evaluate the recommendation system using appropriate metrics such as precision, recall, and F1-score

from sklearn.metrics import precision_score, recall_score, f1_score

def evaluate_recommendations(recommended_items, ground_truth_items):
  """
    Evaluates the recommendation list against a ground truth set of liked items.

    Args:
      recommended_items (list or set): A list or set of items recommended by the system.
      ground_truth_items (list or set): A list or set of items the user actually liked (ground truth).

    Returns:
      dict: A dictionary containing precision, recall, and F1-score.
            Returns None if either recommended_items or ground_truth_items is empty.
  """
  if not recommended_items or not ground_truth_items:
    print("Warning: Cannot compute metrics with empty recommendation list or ground truth.")
    return None

  # Convert to sets for efficient intersection
  recommended_set = set(recommended_items)
  ground_truth_set = set(ground_truth_items)

  # Calculate True Positives (TP) - items recommended and also in ground truth
  true_positives = len(recommended_set.intersection(ground_truth_set))

  # Calculate False Positives (FP) - items recommended but not in ground truth
  false_positives = len(recommended_set - ground_truth_set)

  # Calculate False Negatives (FN) - items in ground truth but not recommended
  false_negatives = len(ground_truth_set - recommended_set)

  # Calculate Precision: TP / (TP + FP)
  precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0.0

  # Calculate Recall: TP / (TP + FN)
  recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0.0

  # Calculate F1-score: 2 * (Precision * Recall) / (Precision + Recall)
  f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0

  return {
      'precision': precision,
      'recall': recall,
      'f1_score': f1
  }


target_anime_name = 'Naruto'

# Generate recommendations for 'Naruto' using the defined function
# We'll use a threshold of 0.7 for this example
recommended_anime_df = recommend_similar_anime_with_threshold(target_anime_name, df, similarity_matrix, similarity_threshold=0.7)
recommended_anime_names = recommended_anime_df['name'].tolist()

print(f"\nRecommended items for '{target_anime_name}':")
print(recommended_anime_names)

naruto_related_anime_df = df[df['name'].str.contains('Naruto', na=False)]
ground_truth_names = naruto_related_anime_df['name'].tolist()

# Let's add a few other well-known shonen anime that might be considered similar
# (These are hypothetical and should be based on actual user data in a real application)
hypothetical_liked_anime = ['One Piece', 'Bleach', 'Fairy Tail', 'Dragon Ball Z']
ground_truth_names.extend(hypothetical_liked_anime)

# Remove duplicates and the target anime itself from the ground truth
ground_truth_names = list(set(ground_truth_names))
if target_anime_name in ground_truth_names:
    ground_truth_names.remove(target_anime_name)

print(f"\nHypothetical Ground Truth for '{target_anime_name}':")
print(ground_truth_names)

# --- Evaluate ---
evaluation_metrics = evaluate_recommendations(recommended_anime_names, ground_truth_names)

if evaluation_metrics:
  print("\nEvaluation Metrics:")
  print(f"Precision: {evaluation_metrics['precision']:.4f}")
  print(f"Recall: {evaluation_metrics['recall']:.4f}")
  print(f"F1-score: {evaluation_metrics['f1_score']:.4f}")

```

**User-based** collaborative filtering recommends items to a user by finding other users with similar preferences.

It compares users based on their ratings or behaviors.

Example: If User A and User B both liked the same shows, and User B liked another anime, recommend that anime to User A.

**Item-based** collaborative filtering, on the other hand, recommends items by comparing items to each other based on how users rated them.

It looks at the similarity between items instead of users.

Example: If many users liked both Anime X and Anime Y, recommend Anime Y to someone who liked Anime X.

**Collaborative filtering** is a recommendation technique that makes predictions based on user-item interactions rather than item content.
It assumes that if users agreed in the past, they will agree again in the future.

It works in two main ways:

**User-based C**F: Finds similar users and recommends items they liked.

**Item-based CF**: Finds similar items based on user ratings and recommends them.

**Steps:**

Create a user-item matrix (users as rows, items as columns).

Measure similarity using cosine similarity, Pearson correlation, etc.

Generate recommendations based on neighbors (users or items).
