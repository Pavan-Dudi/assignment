```python
frequent_itemsets = apriori(data, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

```python
def format_itemset(itemset):
    return ', '.join(sorted(list(itemset)))
```

```python
rules['antecedents_formatted'] = rules['antecedents'].apply(format_itemset)
rules['consequents_formatted'] = rules['consequents'].apply(format_itemset)
rules
```

```python
print("Top 10 Association Rules:")
rules[['antecedents_formatted', 'consequents_formatted', 'support', 'confidence', 'lift']]
```

# **Interview Questions:**

**1.	What is lift and why is it important in Association rules?**

Ans:-Lift in association rules measures how much more likely items are bought together compared to if they were independent. It's important because it highlights truly interesting associations beyond just popular items, helping identify valuable patterns for recommendations or promotions.

**2.	What is support and Confidence. How do you calculate them?**

Ans:- Support is the frequency of an itemset in the dataset (transactions containing the itemset / total transactions). Confidence for a rule X -> Y is the probability that a customer who buys X also buys Y (Support(X U Y) / Support(X)). Support indicates popularity, while confidence measures the rule's reliability.

**3.	What are some limitations or challenges of Association rules mining?**

Association rule mining can generate too many rules, making interpretation difficult. It also doesn't imply causality, only co-occurrence, and can be computationally expensive for large datasets.
