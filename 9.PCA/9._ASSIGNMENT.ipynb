## **Dimensionality Reduction with PCA**

```python
from sklearn.preprocessing import StandardScaler


X = df.values
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(scaler.mean_)
print(scaler.scale_)
print(X_scaled)
```

```python
from sklearn.decomposition import PCA


pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# Calculate cumulative explained variance
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
print(explained_variance_ratio)
print(cumulative_explained_variance)
```

```python
# Plot cumulative explained variance
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.show()
```

```python
# Replace 'optimal_n_components' with the number you determined
optimal_n_components = 2  

pca_final = PCA(n_components=optimal_n_components)
X_pca_final = pca_final.fit_transform(X_scaled)
print(pca_final.explained_variance_ratio_)
print(pca_final.singular_values_)
```

# Clustering with Original Data:

```python
from sklearn.cluster import KMeans

# Apply K-means to the original data
kmeans_original = KMeans(n_clusters=3, random_state=42, n_init=10) 
kmeans_original.fit(X_scaled) 
clusters_original = kmeans_original.labels_
print("Cluster assignments for original data:")
print(clusters_original)
```

```python
# Add the cluster assignments to the original dataframe
df['cluster_original'] = clusters_original
print(df)
```

```python
# we can analyze the characteristics of each cluster in the original feature space
cluster_analysis_original = df.groupby('cluster_original').mean()
print("\nCluster characteristics (original data):")
cluster_analysis_original
```

```python
# Visualize the clusters from the original data
plt.figure(figsize=(10, 8))

sns.scatterplot(x=X_pca_final[:, 0], y=X_pca_final[:, 1], hue=df['cluster_original'], palette='viridis')

plt.title('K-means Clustering on PCA-Reduced Data')
plt.xlabel('Principal Component 1') # Or your chosen original feature names
plt.ylabel('Principal Component 2') # Or your chosen original feature names
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
# Evaluate clustering on original data
silhouette_original = silhouette_score(X_scaled, clusters_original)
davies_bouldin_original = davies_bouldin_score(X_scaled, clusters_original)

print(f"\nClustering Performance on Original Data:")
print(f"Silhouette Score: {silhouette_original:.4f}")
print(f"Davies-Bouldin Index: {davies_bouldin_original:.4f}")
```

```python
# Apply K-means to the PCA-reduced data
kmeans_pca = KMeans(n_clusters=3, random_state=42, n_init=10) # Example: 3 clusters
kmeans_pca.fit(X_pca_final)
clusters_pca = kmeans_pca.labels_
print("\nCluster assignments for PCA-reduced data:")
print(clusters_pca)
```

```python
# Analyze the characteristics of each cluster in the original feature space using PCA clusters
cluster_analysis_pca = df.groupby('cluster_pca').mean()
print("\nCluster characteristics (PCA-reduced data):")
cluster_analysis_pca
```

```python
# Evaluate clustering on PCA-reduced data
silhouette_pca = silhouette_score(X_pca_final, clusters_pca)
davies_bouldin_pca = davies_bouldin_score(X_pca_final, clusters_pca)

print(f"\nClustering Performance on PCA-Reduced Data:")
print(f"Silhouette Score: {silhouette_pca:.4f}")
print(f"Davies-Bouldin Index: {davies_bouldin_pca:.4f}")
```

## **Comparison and Analysis**

```python
# Print comparison and analysis
print("\n--- Comparison and Analysis ---")
print("\nClustering Performance Comparison:")
print(f"  Original Data:")
print(f"    Silhouette Score: {silhouette_original:.4f}")
print(f"    Davies-Bouldin Index: {davies_bouldin_original:.4f}")
print(f"\n  PCA-Reduced Data:")
print(f"    Silhouette Score: {silhouette_pca:.4f}")
print(f"    Davies-Bouldin Index: {davies_bouldin_pca:.4f}")
```

**Discussion:**
- The Silhouette Score is higher for the PCA-reduced data (0.6024) compared to the original data (0.3077). A higher Silhouette Score indicates better-defined clusters with greater separation and less overlap.
- The Davies-Bouldin Index is lower for the PCA-reduced data (0.5397) compared to the original data (1.3164). A lower Davies-Bouldin Index suggests that the clusters are more separated and the points within clusters are more similar.
- Based on these metrics, K-means clustering performed better on the PCA-reduced data.
- PCA helped in reducing the dimensionality while preserving enough information to form more distinct clusters, likely by removing noise or less informative features.
- The visualizations also show clearer separation between clusters in the PCA-reduced space compared to trying to visualize in the high-dimensional original space (which we approximated by using PC1 vs PC2 for the original data visualization as well).


**Discussion on Cluster Characteristics:**
- We can compare the mean values of the original features for the clusters found on the original data versus the clusters found on the PCA-reduced data.
- Ideally, the clusters found by PCA should correspond closely to the clusters found on the original data, but PCA might group instances slightly differently, potentially revealing underlying patterns more effectively.
- By examining the average values of features like 'Alcohol', 'Malic acid', 'Ash', etc., within each cluster from both analyses, we can understand how the clusters differ in terms of the original measurements and whether PCA led to clusters with more interpretable or distinct characteristics.
- For instance, comparing the mean 'Alcohol' content in cluster 0 from the original analysis to the mean 'Alcohol' content in cluster 0 from the PCA analysis can show similarities or differences in how instances with high/low alcohol were grouped.
- The specific interpretation of which cluster corresponds to which depends on the random initialization of K-means, but the overall separation and the characteristics of the groups formed are the key comparison points.

**PCA for Visualization and Dimensionality Reduction:** PCA reduces the number of features while retaining most of the important information. This is practical for visualizing high-dimensional data (e.g., on a 2D scatter plot using principal components) and can improve the efficiency of subsequent analyses by working with fewer dimensions.

**Clustering for Grouping and Insight:** Clustering algorithms like K-means group similar data points together. Applied to original or PCA-reduced data, clustering helps identify natural groupings within the dataset, providing insights into underlying structures and characteristics of different subgroups of data points.

**Exploratory Data Analysis (EDA)**: Use always at the start to understand data basics, find issues (missing values, duplicates), and visualize distributions/relationships.

**Dimensionality Reduction (PCA):** Use when dealing with many features to simplify data, remove noise, and potentially improve performance of subsequent tasks like clustering, especially if EDA suggests high correlation.

**Clustering:** Use when you want to group similar data points without prior labels. Apply after necessary preprocessing like scaling and potentially PCA if it enhances cluster separation and interpretability.
